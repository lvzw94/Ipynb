## 数据处理

### 空值填充

```python
# 统计每一列数据类型
NA.Var
```

```python
df['AASN'].dtypes
```

```python
# 写一个循环
temp=[]
for i in NA.Var:
    temp.append(df[i].dtypes)
```

```python
NA['数据类型']=temp
```

```python
NA
```

空值填充策略：

分类变量：通常分类水平出现的次数多，出现的概率就是最高的。用众数来填充。

数值型变量：幸福指数  收入所处排名都是分类变量

```python
# 去掉年龄
NA[NA.Var!='age'].Var
```

```python
# 对列名进行遍历，依次进行填充
df['AASN'].mode()[0]
```

```python
for i in NA[NA.Var!='age'].Var:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
# 年龄采用均值填充
df.age.fillna(df.age.mean(),inplace=True)
```

```python
df.isnull().sum()
```

### 变量编码

```python
# 首先把用户ID删除
del df['KBM_INDV_ID']
```

```python
df.head()
```

```python
df_object = df.select_dtypes('object')
df_object.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder  # 对特征数值型编码
```

```python
df_object = OrdinalEncoder().fit_transform(df_object)
```

```python
df_object
```

```python
df_copy = df.copy()

# 误操作还原数据
# df = df_copy.copy()
```

```python
# 进行数据的编码转换
for i in df.columns:
    if df[i].dtypes=='object':
        df[i]=OrdinalEncoder().fit_transform(df[[i]])
```

```python
# 统计一下有多少2分类，多少2分类以上的
two = 0
over_two =0
for i in df:
    if df[i].dtypes=='object':
        if df[i].nunique()==2:
            two+=1
        else:
            over_two+=1
            
print(two)
print(over_two)
```

```python
df['GEND'].unique()
```

```python
# map enumerate
df['GEND'].map(dict([(j,i) for i,j in enumerate(df['GEND'].unique())]))
```

```python
columns = []
for i in df:
    if df[i].dtypes=='object':
        if df[i].nunique()==2:
            df[i]=df[i].map(dict([(j,i) for i,j in enumerate(df[i].unique())]))
        else:
            columns.append(i)
```

```python
columns
```

```python
# 转换哑变量
dummies = pd.get_dummies(df[columns],drop_first=True)
```

```python
dummies.shape
```

```python
df.shape
```

```python
df2 = df.drop(columns=columns)
```

```python
df2.shape
```

```python
df_final = pd.concat([df2,dummies],axis=1,sort=False)
df_final.shape
```

## 建模

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
df = df_final.copy()
```

```python
df.head()
```

```python
# for i in df.columns:
#     if df[i].dtypes=='object':
#         df[i]=OrdinalEncoder().fit_transform(df[[i]])
```

```python
X = df.iloc[:,1:]
y = df['resp_flag']
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,y,test_size=0.3,
                                          random_state=420)
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(Xtrain,Ytrain)
clf.score(Xtest,Ytest)
```

### 尝试使用最小叶节点样本数量和最小分割样本数量进行调参

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'min_samples_leaf':list(range(1000,6000,100)),
             'min_samples_split':list(range(4000,6000,100))}
GR = GridSearchCV(tree.DecisionTreeClassifier(),param_grid,cv=5)
GR.fit(Xtrain,Ytrain)

GR.best_params_  # leaf 1500, split 4000
```

```python
clf = tree.DecisionTreeClassifier(criterion='gini',max_depth=4,
                                 min_samples_leaf=1500,
                                 min_samples_split=4000)
clf.fit(Xtrain,Ytrain)
clf.score(Xtest,Ytest)
```

```python
import graphviz
```

```python
features = list(df.columns[1:])
```

```python
dot_data = tree.export_graphviz(clf,feature_names=features,
                               class_names=['NP','P'],
                               filled=True,
                               rounded=True)
```

```python
graph = graphviz.Source(dot_data)
graph
```

### 对数据进行PCA压缩

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=20)
```

```python
Xtrain.shape
```

```python
Xtrain_pca = pca.fit_transform(Xtrain)
```

```python
Xtrain_pca.shape
```

```python
list(pca.explained_variance_ratio_)
```

```python
(pca.explained_variance_ratio_).sum()
```

```python
pd.DataFrame(Xtrain_pca).columns
```

```python
Xtest_pca=pca.transform(Xtest)
```

```python
clf = tree.DecisionTreeClassifier(criterion='gini',max_depth=4,
                                 min_samples_leaf=1500,
                                 min_samples_split=4000)
clf.fit(Xtrain_pca,Ytrain)
clf.score(Xtest_pca,Ytest)
```
